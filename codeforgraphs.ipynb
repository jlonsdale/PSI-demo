```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the log.txt file
data = pd.read_csv('logtimevn.txt')

# Extract relevant columns
runtime = data['time']
n_values = data['n']

# Plotting the chart
plt.figure(figsize=(10, 5))
plt.plot(n_values, runtime, marker='x', linestyle='-', color='b')
plt.title("Runtime vs App Userbase Size (User's Contacts Size=50)")
plt.xlabel('Set Size')
plt.ylabel('Runtime/Nanoseconds')
plt.grid()
plt.xticks(n_values.unique())
plt.ylim(bottom=0, top=3e8)  # Set the y-axis to start at 0 and max at 3*10^8
plt.savefig('runtime_vs_n.png', format='png', dpi=300)  # High DPI for better quality
plt.show()
```
